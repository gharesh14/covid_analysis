```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
```

```python
data_bc  = pd.read_csv('covid_19_data_.csv')
data = data_bc.drop('Province/State', axis = 1)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cont_encoded = le.fit_transform(data['Country'])
data['Country'] = cont_encoded
data.head(2000)
```

```python
date_encoded = le.fit_transform(data['Date'])
data['Date'] = date_encoded
data
```

```python
data['Date'].dtype
```

```python
x = data.drop('Recovered', axis = 1)
x
```

```python
y = data.iloc[:,-3]
y
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
rf_model = RandomForestRegressor()
rf_model.fit(Xtrain, ytrain)
```

```python
y_pred = rf_model.predict(Xtest)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


mae = mean_absolute_error(ytest, y_pred)
mse = mean_squared_error(ytest, y_pred)
r2 = r2_score(ytest, y_pred)


print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
from sklearn.metrics import accuracy_score

y_pred_class = np.round(y_pred)

accuracy = accuracy_score(ytest, y_pred_class)
print("Accuracy:", accuracy)
```
