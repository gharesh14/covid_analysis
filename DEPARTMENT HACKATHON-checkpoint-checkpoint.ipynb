```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
```

```python
data_bc  = pd.read_csv('covid19.csv')
data = data_bc.drop('Province/State', axis = 1)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cont_encoded = le.fit_transform(data['Country'])
data['Country'] = cont_encoded
data.head(2000)
```

```python
date_encoded = le.fit_transform(data['Date'])
data['Date'] = date_encoded
data
```

```python
data['Date'].dtype
```

```python
x = data.drop('Recovered', axis = 1)
x
```

```python
y = data.iloc[:,-3]
y
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
rf_model = RandomForestRegressor()
rf_model.fit(Xtrain, ytrain)
```

```python

y_pred = rf_model.predict(Xtest)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


mae = mean_absolute_error(ytest, y_pred)
mse = mean_squared_error(ytest, y_pred)
r2 = r2_score(ytest, y_pred)


print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
from sklearn.metrics import accuracy_score

y_pred_class = np.round(y_pred)

accuracy = accuracy_score(ytest, y_pred_class)
print("Accuracy:", accuracy)
```

# LONG TERM TREND ANALYSIS

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
covid_data = pd.read_csv('covid19.csv')
```

```python
covid_data['Date'] = pd.to_datetime(covid_data['Date'])
```

```python
daily_cases = covid_data.groupby('Date')['Confirmed'].sum().reset_index()
```

```python
daily_cases['ordinal_date'] = daily_cases['Date'].apply(lambda x: x.toordinal())
```

```python
X = daily_cases[['ordinal_date']]
y = daily_cases['Confirmed']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
future_dates = pd.date_range(start=daily_cases['Date'].max(), periods=365)
future_dates_ordinal = future_dates.to_series().apply(lambda x: x.toordinal())
future_predictions = model.predict(future_dates_ordinal.values.reshape(-1, 1))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(daily_cases['Date'], daily_cases['Confirmed'], label='Actual Cases')
plt.plot(future_dates, future_predictions, label='Predicted Trend', linestyle='--', color='red')
plt.title('COVID-19 Long-Term Trend Analysis')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.legend()
plt.grid(True)
plt.show()
```

# PREDICTING FOR FUTURE

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
covid_data = pd.read_csv('covid19.csv')
```

```python
covid_data['Date'] = pd.to_datetime(covid_data['Date'])
```

```python
covid_data.set_index('Date', inplace=True)
```

```python
daily_cases = covid_data['Confirmed'].resample('D').sum()
```

```python
model = ARIMA(daily_cases, order=(5,1,0))
arima_result = model.fit()
```

```python
forecast_steps = 365  # Forecast for the next 365 days (current year)
forecast = arima_result.forecast(steps=forecast_steps)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(daily_cases, label='Observed')
plt.plot(forecast, label='Forecast', linestyle='--', color='red')
plt.title('COVID-19 Outbreak Forecast for the Current Year')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.legend()
plt.grid(True)
plt.show()
```
